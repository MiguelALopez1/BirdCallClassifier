```python
import os
from pydub import AudioSegment

# Define paths
dataset_path = 'Data/labeledData/Bananaquit'
wav_output_path = 'Data/labeledData/Bananaquit_wav'

# Create a directory to save the converted WAV files
os.makedirs(wav_output_path, exist_ok=True)

# Convert MP3 files to WAV
for file_name in os.listdir(dataset_path):
    if file_name.endswith('.mp3'):
        mp3_path = os.path.join(dataset_path, file_name)
        wav_path = os.path.join(wav_output_path, os.path.splitext(file_name)[0] + '.wav')

        try:
            audio = AudioSegment.from_mp3(mp3_path)
            audio.export(wav_path, format='wav')
            print(f'Successfully converted {file_name}')
        except Exception as e:
            print(f'Failed to convert {file_name}: {e}')

print('Conversion complete.')

# Verify the conversion
print('Converted WAV files:')
print(os.listdir(wav_output_path))
```
